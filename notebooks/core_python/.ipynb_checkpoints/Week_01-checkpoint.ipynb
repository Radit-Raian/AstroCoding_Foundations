### Week 01: Welcome and Setting Up

For the initial task, we will learn how to calculate the distances to the stars by simple appetizers of astronomy coding. In this case we will use the well-known textbook formula- $ d(\mathrm{pc}) = \frac{1}{p (\mathrm{arcseconds})} $. For now, we will work on the Alpha Centauri system. Therefore, we got the parallax for the Aplha Centauri as 747.1 milliarcseconds with an error of 1.2. Let's calculate our intended distance. Later, we will find the semi major axis of the orbit of its two planets named Proxima Centauri b and d, respectively with the famous Kepler's 3rd law- $$ P^2 = \frac{4\pi^2a^3}{GM} $$ where every units are measured in the SI.

```python
parallax_mas = 747.17
parallax_mas_err = 0.61

parallax_as = parallax_mas/1000
parallax_as_err = parallax_mas_err/1000

distance_pc = 1/parallax_as
distance_pc_err = parallax_as_err/parallax_as**2

print(f"The distance from the Earth to the Alpha Centauri system is {distance_pc:.3f} \u00B1 {distance_pc_err:.3f}")
```

```python
import math

period_b = 11.2 * 24 * 3600
period_d = 5.2 * 24 * 3600

M = 0.1221 * 2e30
G = 6.67e-11

distance_b = (G * M * period_b**2 / (4 * math.pi**2))**(1/3)
distance_d = (G * M * period_d**2 / (4 * math.pi**2))**(1/3)

dis_b = (distance_b)/(1.496e11)
dis_d = (distance_d)/(1.496e11)


print(f"The distance from the planets are {dis_b:.3e} and {dis_d:.3e} AU respectively")
```
